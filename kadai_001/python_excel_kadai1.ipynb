```python
#２章
#Excelファイルの作成
import openpyxl

# openpyxl.stylesモジュールのFontクラスをインポート
from openpyxl.styles import Font

workbook = openpyxl.Workbook()

# datetimeモジュールのdatetimeクラスをインポートする
from datetime import datetime

# 現在の日付（年月日）を取得して、変数todayに代入する
today = datetime.today()

#アクティブなシートを取得する
ws = workbook.active

# サイズ設定
font = Font(size=9)


#セルに値を書き込む
ws['B2'] = '請求書'
ws['B4'] = '株式会社ABC'

ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B5'].font = font
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B6'].font = font
ws['B7'] = '担当者名:鈴木一郎 様'
ws['F4'] = 'No.'
ws['F5'] = '日付'
ws['G4'] = '0001'
ws['G5'] = today.strftime('%Y/%m/%d')

# 複数行のデータをリストdataに格納
data = [
    ['商品A', 2, 10000, 20000],
    ['商品B', 1, 15000, 15000],
    ]

# 系列データのタイトルを1行目に設定
ws.cell(row=10, column=2, value='商品名')
ws.cell(row=10, column=3, value='数量')
ws.cell(row=10, column=4, value='単価')
ws.cell(row=10, column=5, value='金額')

# データを追加する開始行を指定
start_row = 11

# 繰り返し処理で各データ項目を正確な位置に設定
for i, row_data in enumerate(data):
    for j, value in enumerate(row_data):
        # 開始行と列のオフセットを考慮してセルに値を設定
        ws.cell(row=start_row + i, column=2 + j, value=value)

#金額合計を'E13'に表示
ws.cell(row=13, column=5, value=(35000) )

# 複数行のデータをリストdata2に格納
data2 = [
    ['小計', 35000],
    ['消費税', 3500],
    ['合計', 38500]
    ]

# データを追加する開始行を指定
start_row = 15

# 繰り返し処理で各データ項目を正確な位置に設定
for i, row_data in enumerate(data2):
    # 2列目に各行の最初の要素（インデックス）を設定
    ws.cell(row=start_row + i, column=2, value=row_data[0])
    # 5列目に各行の2番目の要素（金額）を設定
    ws.cell(row=start_row + i, column=5, value=row_data[1])

# 日付のtodayオブジェクトを、特定のフォーマットで文字列に変換する
filename = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

# Excelファイルを上書き保存する
workbook.save(filename)
```
